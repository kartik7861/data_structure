```python
#Discuss string slicing and provide examples?

# String slicing in Python is a powerful feature that allows you to extract a portion of a string by specifying a start and end index.
 #It uses the syntax string[start:end:step]
 
 #Example 1   
str1 = "Hello, World!"

# Extracting "Hello"
ext1= str1[0:5]              # 0 is the starting index and 5 is the ending index and ending index is excluded
print(ext1)  # Output: Hello

# Extracting "World"          negitive indexing
ext2 = str1[-6:-1]        # -6 is the starting index and -1 is the ending index and ending index is excluded
print(ext2)  # Output: World

#if you skip start and end in the index , python uses default values
ext3 = str1[:4]          # here as i have not given the start index python by default takes 0 as a value
print(ext3)  # output Hell

#The step argument allows you to skip characters:
# extracting every second character
ext4 = str1[::2]      # extract words at a step of 2
print(ext4)  # Output: Hlo ol!

# Slicing in reverse
ext5 = str1[::-1]         
print(ext5)  # Output: !dlroW ,olleH
```

```python
# 3.Describe how to access modify and delete elements in a list with examples ?

#*You can access elements in a list using their index no 
lis =[20,40,50,"ajay","bijay",2.5]           #index stars from 0 
print(lis(1))       #output = 40 2nd element of list
print(lis(-1))      #output = 2.5 last element of list


#**You can change the value of an element at a specific index using assignment 

lis[1]= 45   #change 2nd element of the list to 45 

lis[-1]= 25   #change last element of the list to 25 

#***you can also add element in the list by using append insert or extend keywords 

lis.append(80)   # adds element 80 in the list 

lis.insert(1,76)   #inserts 76 at index 1

lis.extend(39,78)  #adds two or more elements at a time




#****you can also delete element in the list by using del() , remove() or pop() keywords 

my_list = [10, 20, 30, 40, 50]

del my_list[1]  # Remove the element at index 1 (20)


my_list.remove(30)  # Remove the first occurrence of 30

# Using pop() to remove and return an element by index
popped_element = my_list.pop(0)  # Remove and return the first element
print(popped_element)  # Output: 10

# Using pop() without an index removes the last element
last_element = my_list.pop()
print(last_element)  # Output: 50
```

```python
# 5. Describe the key features of sets and provide examples of their use ?

#Sets represents an unordered collection of unique elements.
# the order of elements in sets is not fixed that means they can be stored anywhere do not gurrantee a specific order means index is not fixed.
# Sets automatically remove duplicate entries. Each element in a set must be unique, that means if yoy write same elements in set the duplicate entries will be removed automatically.
# sets are mutable, means you can add or remove elements after the set has been created.
# sets support mathematical operations like union, intersection, difference, and symmetric difference.

set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Union
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5}

# Intersection
print(set_a & set_b)  # Output: {3}

# Difference
print(set_a - set_b)  # Output: {1, 2}

# Symmetric Difference
print(set_a ^ set_b)  # Output: {1, 2, 4, 5}

```

```python
#7.Describe how to add modify and delete items in a dictionary with examples ?


#dictionaries are mutable, unordered collections of key-value pairs.

# Creating a dictionary
my_dict = {'name': 'Ajay', 'age': 25}

# Adding a new item
my_dict['city'] = 'jaipur'
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Modifying item
my_dict['age'] = 26    # here age will be change from 25 to 26
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}

my_dict['city'] = 'chandighar' # here name of the city will change from jaipur to chandighar
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'}

# Deleting an item using del
del my_dict['city']                   #city will be deleted from dictionary along with its values 
print(my_dict)  # Output: {'name': 'Alice', 'age': 26}
```
